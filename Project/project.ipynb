# Data Load

This section loads the required data into the notebook for analysis and processing.

- Import necessary libraries
- Read the source data
- Validate the loaded dataset
- Prepare the data for downstream steps

```python
# import 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Dataset/modified/DSE_Data_Manupulation_with_Sector.csv')
df.head()
```

```python
print(df.info())
print("\nMissing values per column:")
display(df.isna().sum())

print("\nDuplicate rows:", df.duplicated().sum())
print("Duplicate timestamps:", df["Date"].duplicated().sum())

print("\nDescriptive statistics:")
display(df.describe().T)
```

```python


# Count each sector, including missing values
sector_counts = (
    df['Sector']
    .fillna('Missing')      # Replace NaN with "Missing"
    .value_counts()
)

plt.figure(figsize=(12, 6))
plt.bar(sector_counts.index, sector_counts.values)

plt.title("Distribution of Sector (Including Missing Values)")
plt.xlabel("Sector")
plt.ylabel("Number of Rows")

plt.xticks(rotation=45, ha='right')

# Add value labels on top of bars
for i, v in enumerate(sector_counts.values):
    plt.text(i, v + sector_counts.max() * 0.005, f"{v:,}",
             ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```

```python

# Count unique Trading_Code per Sector
sector_counts = (
    df[['Trading_Code', 'Sector']]
    .drop_duplicates(subset='Trading_Code')   # Count each Trading_Code only once
    .fillna({'Sector': 'Missing'})
    .groupby('Sector')['Trading_Code']
    .count()
    .sort_values(ascending=False)
)

plt.figure(figsize=(12, 6))
plt.bar(sector_counts.index, sector_counts.values)

plt.title("Number of Unique Trading Codes by Sector")
plt.xlabel("Sector")
plt.ylabel("Unique Trading Codes")

plt.xticks(rotation=45, ha='right')

# Add value labels
for i, v in enumerate(sector_counts.values):
    plt.text(i, v + 0.5, str(v), ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```
